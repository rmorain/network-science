Compute the degree centrality, eigenvector centrality, and Katz centrality for the following graphs:



```python
import networkx as nx
```

```python
# Kite graph
kite_graph = nx.Graph({0:[1], 1:[2], 2:[3, 5], 3:[2, 4], 4:[3,5], 5:[4,2]})
nx.draw(kite_graph, with_labels=True)
```

```python
print("Degree Centrality: ", nx.degree_centrality(kite_graph))
print("Eigenvector Centrality", nx.eigenvector_centrality(kite_graph))
print("Katz Centrality", nx.katz_centrality(kite_graph))
```

```python
barbell_graph = nx.Graph({
    0:[1,2,3],
    1:[2,3,0],
    2:[3,0,1],
    3:[2,1,0,4],
    4:[3,5],
    5:[6,4],
    6:[5, 9,8,7],
    7:[6,8,9],
    8:[7,9,6],
    9:[8,7,6],
})
nx.draw(barbell_graph, with_labels=True)
```

```python
print("Degree Centrality: ", nx.degree_centrality(barbell_graph))
print("Eigenvector Centrality", nx.eigenvector_centrality(barbell_graph))
print("Katz Centrality", nx.katz_centrality(barbell_graph))
```

```python
spoke_graph = nx.Graph({
    1:[2, 3],
    2:[1, 3],
    3:[1,2],
    4:[6, 5, 10],
    5:[6, 4],
    6:[5,4],
    7:[10,9,8],
    8:[9,7],
    9:[8,7],
    10:[4,7,1]
})
nx.draw(spoke_graph, with_labels=True)
```

```python
print("Degree Centrality: ", nx.degree_centrality(spoke_graph))
print("Eigenvector Centrality", nx.eigenvector_centrality(spoke_graph))
print("Katz Centrality", nx.katz_centrality(spoke_graph))
```

### Discuss which centrality measure would be most useful for stopping the spread of simple contagions in terms of which edges you'd cut from the graph



The most useful centrality measure for stopping the spread of a simple contagion would be either Eigenvector centrality or Katz centrality. In these examples, both of these measures seem to identify the best edge to remove is in between the nodes with highest and that nodes most central neighbor. Because these are undirected graphs and Katz centrality is designed for directed graphs, I would favor Eigenvector centrality.
